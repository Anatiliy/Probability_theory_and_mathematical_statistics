Задача 3. На соревновании по биатлону один из трех спортсменов стреляет и попадает в мишень. Вероятность попадания для первого спортсмена равна 0.9, для второго — 0.8, для третьего — 0.6. Найти вероятность того, что выстрел произведен: a). первым спортсменом б). вторым спортсменом в). третьим спортсменом.

```python
# вероятность того, что спортсмен 1 попал в мишень, а остальные промахнулись
result1 = 0.9 * (1 - 0.8) * (1 - 0.6)
# вероятность того, что спортсмен 2 попал в мишень, а остальные промахнулись
result2 = 0.8 * (1 - 0.9) * (1 - 0.6)
# вероятность того, что спортсмен 3 попал в мишень, а остальные промахнулись
result3 = 0.6 * (1 - 0.8) * (1 - 0.9)
# вероятность того, что только один спортсмен попал в мишень, а остальные промахнулись
result = result1 + result2 + result3
print('вероятность того, что выстрел произведен первым спортсменом =', result1 / result)
print('вероятность того, что выстрел произведен вторым спортсменом =', result2 / result)
print('вероятность того, что выстрел произведен третьим спортсменом =', result3 / result)
```

```python
# вероятность того, что стрелять будет спортсмен 1 и попадёт в мишень
result1 = 0.9 * (1 / 3)
# вероятность того, что стрелять будет спортсмен 2 и попадёт в мишень
result2 = 0.8 * (1 / 3)
# вероятность того, что стрелять будет спортсмен 3 и попадёт в мишень
result3 = 0.6 * (1 / 3)
# вероятность того, что мишень будет поражена
result = result1 + result2 + result3
print('вероятность того, что выстрел произведен первым спортсменом =', result1 / result)
print('вероятность того, что выстрел произведен вторым спортсменом =', result2 / result)
print('вероятность того, что выстрел произведен третьим спортсменом =', result3 / result)
```

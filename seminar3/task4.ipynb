Задача 4. В университет на факультеты A и B поступило равное количество студентов, а на факультет C студентов поступило столько же, сколько на A и B вместе. Вероятность того, что студент факультета A сдаст первую сессию, равна 0.8. Для студента факультета B эта вероятность равна 0.7, а для студента факультета C - 0.9. Студент сдал первую сессию. Какова вероятность, что он учится: a). на факультете A б). на факультете B в). на факультете C?

```python
# вероятность, что студент учится на факультете A
chance_A = 1 / 4
# вероятность, что студент учится на факультете B
chance_B = 1 / 4
# вероятность, что студент учится на факультете C
chance_C = 2 / 4
```

```python
# вероятность того, что студент сдаст первую сессию, и учиться на факультете A
result1 = chance_A * 0.8
# вероятность того, что студент сдаст первую сессию, и учиться на факультете B
result2 = chance_B *  0.7
# вероятность того, что студент сдаст первую сессию, и учиться на факультете C
result3 = chance_C * 0.9
# вероятность того, что студент сдаст первую сессию
result = result1 + result2 + result3
print('вероятность того, что студент, сдавший первую сессию, учиться на факультете A =', result1 / result)
print('вероятность того, что студент, сдавший первую сессию, учиться на факультете B =', result2 / result)
print('вероятность того, что студент, сдавший первую сессию, учиться на факультете C =', result3 / result)
```
